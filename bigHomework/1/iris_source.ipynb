```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import util
%matplotlib inline
```

```python
datMat = []
labels = []
result = []
with open('iris.csv', 'rb') as f:
    line = f.readlines()
    labels = line[0].strip().split(',')
    for x in line[1:]:
        x = x.strip()
        res = x.split(',')
        datMat.append([float(i) for i in res[:4]])
        result.append(res[4])
t = lambda x: {'setosa': 0, 'versicolor': 1, 'virginica': 2}[x]
result = map(t, result)
print result
```

```python
df = pd.DataFrame(datMat)
df['class'] = result
pd.scatter_matrix(df.iloc[:,:-1], figsize=(9, 9), \
                  diagonal='kde', marker='o', s=40, alpha=.4, c=df['class'])
```
