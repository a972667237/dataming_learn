```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import util
import time
%matplotlib inline
```

```python
# load data
datMat = []
labels = []
result = []
with open('iris.csv', 'rb') as f:
    line = f.readlines()
    labels = line[0].strip().split(',')
    for x in line[1:]:
        x = x.strip()
        res = x.split(',')
        datMat.append([float(i) for i in res[:4]])
        result.append(res[4])
t = lambda x: {'setosa': 0, 'versicolor': 1, 'virginica': 2}[x]
result = map(t, result)
print datMat[:3]
print result[:3]
print labels[:3]
```

```python
# print origin pic
df = pd.DataFrame(datMat)
df['class'] = result
pd.scatter_matrix(df.iloc[:,:-1], figsize=(9, 9), \
                  diagonal='kde', marker='o', s=40, alpha=.4, c=df['class'])
```

```python
# init kmean
kmeans_pred = []
kmeans_score = []
kmeans_time = []
kmeans_data = []
```

```python
score_arr = []
s = time.time()
for i in xrange(0, 50):
    pred, score = util.do_kmeans(datMat, result, 3)
    score_arr.append(score)
kmeans_time.append(time.time()-s)
kmeans_pred.append(pred)
kmeans_score.append([sum([i[0] for i in score_arr])/len(score_arr), sum([i[1] for i in score_arr])/len(score_arr)])
kmeans_data.append(score_arr)
print kmeans_time[-1], kmeans_score[-1]
```
