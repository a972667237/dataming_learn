## 毒蘑菇例子

```python
"""
毒蘑菇例子
原始数据下载地址：http://archive.ics.uci.edu/ml/machine-learning-databases/mushroom/
"""
mushDatSet = [line.strip().split(',') for line in open('agaricus-lepiota_new.data').readlines()]
# 前三行数据
print(mushDatSet[:3])
```

```python
L,suppData=apriori.apriori(mushDatSet, minSupport=0.46)
```

```python
# 目标项包含'0p'的关系
freq_item_list = []
supp_list = []
for itemset_list in L:
    for itemset in itemset_list:
        if itemset.intersection({'0p'}):
            freq_item_list.append(itemset)
            supp_list.append(suppData[itemset])
```

```python
import numpy as np
# 目标项包含{'0p'}且支持度前10大的关系
for index in np.argsort(supp_list)[-10:]:
    item = freq_item_list[index]
    print item, supp_list[index]
```

```python
np.argsort(supp_list)[:-1:]
```

```python
# 包含毒蘑菇‘0p’的频繁项集
for itemset_list in L:
    for itemset in itemset_list:
        if itemset.intersection({'0p'}): print itemset
```

```python
rules=apriori.generateRules(L,suppData, minConf=0.3)
```

```python
len(rules)
```

```python
# 目标项是{'0p'}的关系
for item1, item2, conf in rules:
    if item2=={'0p'}: print item1, '-->', item2, ' conf: ', conf
```

```python
# 目标项包含'0p'的关系
rule_list = []
conf_list = []
for item1, item2, conf in rules:
    if item2.intersection({'0p'}): 
        print item1, '-->', item2, ' conf: ', conf
        rule_list.append((item1, item2))
        conf_list.append(conf)
```

```python
import numpy as np
# 目标项包含{'0p'}且置信度前10大的关系
for index in np.argsort(conf_list)[-10:]:
    item1, item2 = rule_list[index]
    print item1, '-->', item2, ' conf: ', conf
```
